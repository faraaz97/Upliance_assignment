```python
import pandas as pd

# Loading datasets
user_details = pd.read_excel('Assignment.xlsx',sheet_name='UserDetails.csv')
cooking_sessions = pd.read_excel('Assignment.xlsx',sheet_name='CookingSessions.csv')
order_details = pd.read_excel('Assignment.xlsx',sheet_name='OrderDetails.csv')

# Displaying datasets
print(user_details)
print(cooking_sessions)
print(order_details)
```

## Data Cleaning

```python
# Handling missing values 
user_details = user_details.dropna()
cooking_sessions = cooking_sessions.dropna() 
order_details = order_details.dropna()

print(user_details)
print(cooking_sessions)
print(order_details)
```

```python
# Ensuring correct data types
user_details['User ID'] = user_details['User ID'].astype(str) 
cooking_sessions['Session ID'] = cooking_sessions['Session ID'].astype(str) 
order_details['Order ID'] = order_details['Order ID'].astype(str)

print(user_details)
print(cooking_sessions)
print(order_details)
```

```python
#Removing Duplicates
user_details.drop_duplicates(inplace=True) 
cooking_sessions.drop_duplicates(inplace=True) 
order_details.drop_duplicates(inplace=True)

print(user_details)
print(cooking_sessions)
print(order_details)
```

## Merging all the datasets

```python
# Merging UserDetails with CookingSessions
merged_data = pd.merge(user_details, cooking_sessions, on='User ID')

# Merging with OrderDetails
final_data = pd.merge(merged_data, order_details, on=['User ID', 'Dish Name', 'Meal Type'])

print(final_data)
```

## Analyzing the relationships

```python
# Number of orders per cooking session
orders_per_session = final_data.groupby('Session ID_x')['Order ID'].count().reset_index()
orders_per_session.columns = ['Session ID', 'Order Count']

# Most orders in a cooking session
max_orders_session = orders_per_session[orders_per_session['Order Count'] == orders_per_session['Order Count'].max()]
print(max_orders_session)
```

## Identifying the popular Dishes

```python
# Most ordered dishes
popular_dishes = final_data['Dish Name'].value_counts().reset_index()
popular_dishes.columns = ['Dish Name', 'Order Count']
print(popular_dishes)
```

## Demographic Analysis

```python
# Example: Number of orders by user age group
final_data['Age Group'] = pd.cut(final_data['Age'], bins=[0, 18, 25, 35, 45, 55, 65, 100], labels=['<18', '18-25', '26-35', '36-45', '46-55', '56-65', '65+'])
orders_by_age = final_data.groupby('Age Group')['Order ID'].count().reset_index()
orders_by_age.columns = ['Age Group', 'Order Count']
print(orders_by_age)
```

## Visualizations

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar chart for popular dishes
plt.figure(figsize=(10, 5))
sns.barplot(x='Dish Name', y='Order Count', data=popular_dishes.head(10))
plt.title('Top 10 Popular Dishes')
plt.xticks(rotation=45)
plt.show()

# Line graph for order trends over time
final_data['Order Date'] = pd.to_datetime(final_data['Order Date'])
orders_over_time = final_data.groupby('Order Date')['Order ID'].count().reset_index()
plt.figure(figsize=(10, 5))
sns.lineplot(x='Order Date', y='Order ID', data=orders_over_time)
plt.title('Order Trends Over Time')
plt.show()

# Pie chart for demographic distribution
plt.figure(figsize=(7, 7))
plt.pie(orders_by_age['Order Count'], labels=orders_by_age['Age Group'], autopct='%1.1f%%')
plt.title('Orders by Age Group')
plt.show()
```
